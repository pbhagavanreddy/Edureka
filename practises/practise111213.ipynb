```python
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_excel('Online_Retail.xlsx')
```

```python
df.head()
```

```python
df['Description']=df['Description'].str.strip()
# df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo']=df['InvoiceNo'].astype('str')
df=df[~df['InvoiceNo'].str.contains('C')]
```

```python
df.head()
```

```python
basket=(df[df['Country']=='France'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0))
basket
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=0:
        return 1
basket_sets=basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets.head()
```

```python
frequent_items=apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_items, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[(rules['lift']>=6) & (rules['confidence']>=0.8)]
```

# Recommended Systems

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df2 = pd.read_csv('Recommend.csv',names=['user_id', 'movie_id', 'rating', 'timestamp'])
df2.head()
```

```python
n_users = df2.user_id.unique().shape[0]
n_movies = df2.movie_id.unique().shape[0]
print(n_users)
print(n_movies)
```

```python
train_data, test_data = train_test_split(df2, test_size=0.25)
```

```python
train_data_matrix = np.zeros((n_users, n_movies))
for line in train_data.itertuples():
    #[user_id index, movie_id index] = given rating.
    train_data_matrix[line[1]-1, line[2]-1] = line[3] 
    
train_data_matrix
```

```python
test_data_matrix = np.zeros((n_users, n_movies))
for line in test_data.itertuples():
    #[user_id index, movie_id index] = given rating.
    test_data_matrix[line[1]-1, line[2]-1] = line[3]
test_data_matrix
```

```python
from sklearn.metrics import pairwise_distances
user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
movie_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')
mean_user_rating = train_data_matrix.mean(axis=1)[:, np.newaxis] 
ratings_diff = (train_data_matrix - mean_user_rating) 
user_pred = mean_user_rating + user_similarity.dot(ratings_diff) / np.array([np.abs(user_similarity).sum(axis=1)]).T
user_pred
```

```python
movie_pred = train_data_matrix.dot(movie_similarity) / np.array([np.abs(movie_similarity).sum(axis=1)])
movie_pred
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(pred, test):
    pred = pred[test.nonzero()].flatten() 
    test = test[test.nonzero()].flatten()
    return sqrt(mean_squared_error(pred, test))
```

```python
rmse(user_pred, test_data_matrix)
```

```python
rmse(movie_pred, test_data_matrix)
```
