```python
train=pd.read_csv('train.csv')
train.describe()
```

```python
gr = train.groupby('Sex')
gr.groups['female']
# train.loc[gr.groups['female']] #OR below gives same output
gr.get_group('female')
```

```python
cl=train.groupby('Sex')
for key,subset in cl:
    print(key, len(subset))
```

```python
res=[]
for key,subset in cl:
    total = len(subset)
    survive = subset['Survived'].sum()
    percent = round((survive/total)*100, 2)
    res.append((key,percent))
```

```python
res
```

```python
res_df=pd.DataFrame(res, columns=['Pclass', 'SurvPercent'])
```

```python
plt.bar(res_df['Pclass'], res_df['SurvPercent'])
plt.show()
```

```python
mult=train.groupby(['Sex','Pclass','Embarked'])
res=[]
for key,subset in mult:
    total = len(subset)
    survive = subset['Survived'].sum()
    percent = round((survive/total)*100, 2)
    res.append((key,percent))


res
```
