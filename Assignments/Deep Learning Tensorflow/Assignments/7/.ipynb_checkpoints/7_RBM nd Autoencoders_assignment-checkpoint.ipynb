```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from keras import datasets
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, BatchNormalization, Activation
from keras.models import Model
from keras.optimizers import Adam
```

```python
# loaded MNIST data1
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()
```

```python
#Define hyper parameters
batch_size = 1000
epochs = 10
```

```python
num_classes = 10
# normalize data
train_images = train_images.astype('float32')
test_images  = test_images.astype('float32')
train_images /= 255
test_images /= 255
```

```python
train_images = train_images.reshape(60000, 28, 28, 1)
test_images = test_images.reshape(10000, 28, 28, 1)
```

```python
print(train_images.shape)
print(test_images.shape)
```

```python
val_images = test_images[:7000]
test_images = test_images[7000:]
```

```python
print(val_images.shape)
print(test_images.shape)
```

```python
#Convolutional AutoEncoder for denoising

input_img = Input(shape=(28, 28, 1))
x = Conv2D(64, (3, 3), padding='same')(input_img)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(16, (3, 3), padding='same')(x)
x = BatchNormalization()(x)
x = Activation('relu')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

x = Conv2D(16, (3, 3), padding='same')(encoded)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), padding='same')(x)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(3, (3, 3), padding='same')(x)
x = BatchNormalization()(x)
decoded = Activation('sigmoid')(x)
```

```python
model = Model(input_img, decoded)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
```

```python
#Train AutoEncoder
history = model.fit(train_images, train_images,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(val_images, val_images))
```

```python
# Evaluate with test data
score = model.evaluate(test_images, test_images, verbose=1)
print(score)
```

```python
# Visualize original image and decoded image
mnist_test = model.predict(test_images)
mnist_val = model.predict(val_images)
```

```python
print("MNIST_test: {0}\nMNIST_val: {1}".format(np.average(mnist_test), np.average(mnist_val)))
```

```python
# definition to show original image and reconstructed image
def showOrigDec(orig, denoise, num=10):
    import matplotlib.pyplot as plt
    n = num
    plt.figure(figsize=(20, 4))

    for i in range(n):
        # display original
        ax = plt.subplot(3, n, i+1)
        plt.imshow(orig[i].reshape(28, 28))
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        
        # display decoded image
        ax = plt.subplot(3, n, i +1 + n + n)
        plt.imshow(denoise[i])
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()
```

```python
showOrigDec(test_images, mnist_test)
```
