```python
import pandas as pd
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
# df = pd.read_fwf('letterCG.bin',widths=None)
```

```python
df=pd.read_csv('letterCG.csv', sep=' ')
df.head(3)
```

```python
df.describe()
```

```python
df.columns
```

```python
features=['x-box', 'y-box', 'width', 'high', 'onpix', 'x-bar', 'y-bar',
       'x2bar', 'y2bar', 'xybar', 'x2ybr', 'xy2br', 'x-ege', 'xegvy', 'y-ege',
       'yegvx']
label=['Class']
X=df[features]
Y=df[label]
seed=100
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.20, random_state=seed)
```

```python
ad=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1), n_estimators=30,
                      learning_rate=1.0, random_state=seed)

model=ad.fit(X_train,Y_train)
pred_val=model.predict(X_test)
accuracy=accuracy_score(Y_test,pred_val)
accuracy
```

#### Repeat step2 with max_depth set as 2.

```python
ad=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=2), n_estimators=30,
                      learning_rate=1.0, random_state=seed)

model=ad.fit(X_train,Y_train)
pred_val=model.predict(X_test)
accuracy=accuracy_score(Y_test,pred_val)
accuracy
```
