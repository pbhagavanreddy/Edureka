```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
dsscore=pd.read_csv('DSScoreTerm1.csv')
mascore=pd.read_csv('MathScoreTerm1.csv')
pyscore=pd.read_csv('PhysicsScoreTerm1.csv')

dsscore.drop(['Name','Ethinicity'], axis=1, inplace=True)
mascore.drop(['Name','Ethinicity'], axis=1, inplace=True)
pyscore.drop(['Name', 'Ethinicity'], axis=1, inplace=True)

print(dsscore.head(3))
print("\n", mascore.head(3))
print("\n", pyscore.head(3))
```

```python
dsscore.fillna(0, inplace=True)
mascore.fillna(0, inplace=True)
pyscore.fillna(0, inplace=True)
print(dsscore.isna().sum())
print("\n", mascore.isna().sum())
print("\n", pyscore.isna().sum())
```

```python
ds_ma = (pd.merge(dsscore,mascore, on=['ID', 'Age', 'Sex'], suffixes=('_DS', '_MT')))
merger = (pd.merge(ds_ma,pyscore, on=['ID', 'Age', 'Sex']))
merger.rename(columns={"Score":"Score_PY", "Subject":"Subject_PY"}, inplace=True)
merger.head()
```

```python
merger = merger[['ID', 'Age', 'Sex', 'Subject_DS','Score_DS','Subject_MT', 'Score_MT','Subject_PY','Score_PY']]
merger.Sex[merger.Sex == 'M'] = 1
merger.Sex[merger.Sex == 'F'] = 2

merger.head()
```

```python
merger.to_csv('ScoreFinal.csv', index=False)
```

# #Enhancements for code

```python
merger.Score_DS[merger.Score_DS==0]=dsscore['Score'].mean()
merger.Score_MT[merger.Score_MT==0]=mascore['Score'].mean()
merger.Score_PY[merger.Score_PY==0]=pyscore['Score'].mean()
```

```python
dsscore2=pd.read_csv('DSScoreTerm1.csv')
subset=dsscore2['Ethinicity']
l=LabelEncoder()
merger['NumericEthnicity']=l.fit_transform(subset)
```
