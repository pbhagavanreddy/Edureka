```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
features=['Avg_Session_Length', 'Time_on_App', 'Time_on_Website', 'Length_of_Membership']
target=['Yearly_Amount_Spent']
X=fyntra[features]
Y=fyntra[target]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.20,random_state=85)
lr=LinearRegression()
model=lr.fit(X_train, Y_train)
pred_values=model.predict(X_test)
pred_values

print("Root Mean Squared Error = ", mean_squared_error(Y_test, pred_values))
print("Mean Absolute Error = ", mean_absolute_error(Y_test, pred_values))
print("\nrandom state number will always get the same output to get reproducible results")
```
