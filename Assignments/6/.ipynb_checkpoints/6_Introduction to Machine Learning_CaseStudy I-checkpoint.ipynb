#### 2. Data Manipulation:

```python
pris['total_benefitted']=pris.iloc[:,2:6].sum(axis=1)
```

```python
pris.head(5)
```

```python
totdata = [['totals', 2013, pris.loc[:,'No. of Inmates benefitted by Elementary Education'].sum(axis=0),
          pris.loc[:,'No. of Inmates benefitted by Adult Education'].sum(axis=0),
          pris.loc[:,'No. of Inmates benefitted by Higher Education'].sum(axis=0),
          pris.loc[:,'No. of Inmates benefitted by Computer Course'].sum(axis=0),
          pris.loc[:,'total_benefitted'].sum(axis=0)]]
```

```python
pris9 = pd.DataFrame(data=totdata, 
                     index=None, 
                     columns=['STATE/UT', 'YEAR', 'No. of Inmates benefitted by Elementary Education', 'No. of Inmates benefitted by Adult Education', 'No. of Inmates benefitted by Higher Education', 'No. of Inmates benefitted by Computer Course', 'total_benefitted'])
```

```python
pris2 = pris.append(pris9, ignore_index=True)
```
